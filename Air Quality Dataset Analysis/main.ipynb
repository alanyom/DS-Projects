Section 2:

The dataset contains 9357 instances of hourly averaged responses from an array of 5 metal oxide chemical sensors embedded in an Air Quality Chemical Multisensor Device. The device was located on the field in a significantly polluted area, at road level,within an Italian city. Data were recorded from March 2004 to February 2005. The data consists of hourly averaged concentrations for CO, Non Metanic Hydrocarbons, Benzene, Total Nitrogen Oxides (NOx) and Nitrogen Dioxide (NO2) and were provided by a co-located reference certified analyzer. Missing values are tagged with -200 value.

Column Description:

CO(GT):
Definition: The true hourly averaged concentration of carbon monoxide (CO) in milligrams per cubic meter (mg/m3).
Source: Reference analyzer.
Purpose: Indicates air pollution levels related to carbon monoxide.

PT08.S1(CO):
Definition: Hourly averaged sensor response of tin oxide, nominally targeted to detect carbon monoxide (CO).
Unit: Sensor-specific response (arbitrary units or normalized values).
Purpose: Proxy for detecting carbon monoxide via sensor technology.

NMHC(GT):
Definition: True hourly averaged concentration of overall Non-Methane Hydrocarbons (NMHC) in micrograms per cubic meter (µg/m3).
Source: Reference analyzer.
Purpose: Tracks hydrocarbon pollution in the air.

C6H6(GT):
Definition: True hourly averaged concentration of benzene (C6H6) in micrograms per cubic meter (µg/m3).
Source: Reference analyzer.
Purpose: Monitors air quality by tracking benzene, a toxic pollutant.

PT08.S2(NMHC):
Definition: Hourly averaged sensor response of titania, nominally targeted to detect Non-Methane Hydrocarbons (NMHC).
Unit: Sensor-specific response.
Purpose: Proxy for hydrocarbon detection via sensor readings.

NOx(GT):
Definition: True hourly averaged concentration of nitrogen oxides (NOx) in parts per billion (ppb).
Source: Reference analyzer.
Purpose: Monitors combined levels of nitrogen monoxide (NO) and nitrogen dioxide (NO2).

PT08.S3(NOx):
Definition: Hourly averaged sensor response of tungsten oxide, nominally targeted to detect nitrogen oxides (NOx).
Unit: Sensor-specific response.
Purpose: Proxy for NOx detection using sensors.

NO2(GT):
Definition: True hourly averaged concentration of nitrogen dioxide (NO2) in micrograms per cubic meter (µg/m3).
Source: Reference analyzer.
Purpose: Tracks NO2 levels specifically, a significant air pollutant impacting human health.



Most of these columns will not be used as I plan to dive deeper into the nitrogen dioxide data as stated in section 1


```python
import numpy as np

# Replace -200 with NaN for easier handling of missing values
df.replace(-200, np.nan, inplace=True)

print("Dataset Shape:")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}\n")

print("Column Names:")
print(df.columns.tolist(), "\n")

print("Descriptive Statistics for NO2(GT):")
desc_stats_no2 = df['NO2(GT)'].describe()
print(desc_stats_no2, "\n")

mean_no2 = df['NO2(GT)'].mean()
median_no2 = df['NO2(GT)'].median()

print("Mean Value for NO2(GT): ")
print(mean_no2)
print("Median Value for NO2(GT): ")
print(median_no2)

print("Missing Values Per Column:")
print(df.isna().sum(), "\n")

no2_pollutants = ['NO2(GT)', 'PT08.S4(NO2)']
print("Correlation Matrix for Nitrogen Dioxide (NO2):")
print(df[no2_pollutants].corr())
```

Statistical Analysis:
Pinpoint which months have significantly higher or lower NO2 levels.

Hypothesis:
Null Hypothesis (H0): There is no significant difference in the average NO2 levels between months.
Alternative Hypothesis (HA): There is a significant difference in the average NO2 levels between at least two months.
